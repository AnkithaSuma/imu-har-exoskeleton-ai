```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Load training data
X_train = pd.read_csv(
    r"D:\imu-har-exoskeleton-ai\data\UCI HAR Dataset\UCI HAR Dataset\train\X_train.txt",
    sep=r"\s+",
    header=None
)

y_train = pd.read_csv(
    r"D:\imu-har-exoskeleton-ai\data\UCI HAR Dataset\UCI HAR Dataset\train\y_train.txt",
    header=None
)

# Load testing data
X_test = pd.read_csv(
    r"D:\imu-har-exoskeleton-ai\data\UCI HAR Dataset\UCI HAR Dataset\test\X_test.txt",
    sep=r"\s+",
    header=None
)

y_test = pd.read_csv(
    r"D:\imu-har-exoskeleton-ai\data\UCI HAR Dataset\UCI HAR Dataset\test\y_test.txt",
    header=None
)
```

```python
# Load feature names
features = pd.read_csv(
    r"D:\imu-har-exoskeleton-ai\data\UCI HAR Dataset\UCI HAR Dataset\features.txt",
    sep=r"\s+",
    header=None,
    names=["Index", "Feature"]
)
```

```python
# Assign feature names
X_train.columns = features["Feature"]
X_test.columns = features["Feature"]
```

```python
# Check shapes
print("X_train:", X_train.shape)
print("y_train:", y_train.shape)

print("X_test:", X_test.shape)
print("y_test:", y_test.shape)
```

```python
activity_labels = {
    1: "WALKING",
    2: "WALKING_UPSTAIRS",
    3: "WALKING_DOWNSTAIRS",
    4: "SITTING",
    5: "STANDING",
    6: "LAYING"
}
```

```python
y_train["Activity"] = y_train[0].map(activity_labels)
y_test["Activity"] = y_test[0].map(activity_labels)
```

```python
y_train.head()
```

```python
label_encoder = LabelEncoder()
```

```python
y_train_encoded = label_encoder.fit_transform(y_train["Activity"])
y_test_encoded = label_encoder.transform(y_test["Activity"])
```

```python
print(y_train_encoded[:10])
```

```python
print(label_encoder.classes_)
```

```python
for i, label in enumerate(label_encoder.classes_):
    print(i, "->", label)
```

```python
X_train_final, X_val, y_train_final, y_val = train_test_split(
    X_train,
    y_train_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_train_encoded
)
```

```python
print("X_train_final:", X_train_final.shape)
print("X_val:", X_val.shape)

print("y_train_final:", y_train_final.shape)
print("y_val:", y_val.shape)
```

```python
print("Final Training Set:", X_train_final.shape)
print("Validation Set:", X_val.shape)
print("Test Set:", X_test.shape)
```
